# EDA & qualitative inspection
Inspect the synthetic/real sample, the resampled products, and model outputs.
Run `tir-prepare-data` first to populate `data/processed/`.

```python
import pandas as pd
from tir.utils.geo import read_raster
from tir.utils.viz import save_panel

df = pd.read_csv('../data/processed/manifest.csv')
print(df['split'].value_counts())
row = df.iloc[0]
lr = read_raster('../' + row['lr_tir']).data
hr = read_raster('../' + row['hr_tir']).data
rgb = read_raster('../' + row['rgb']).data
print('LR', lr.shape, 'HR', hr.shape, 'RGB', rgb.shape)
save_panel([lr, hr, rgb], ['LR TIR 200m', 'HR TIR 100m', 'RGB 100m'], 'eda_panel.png')
```
